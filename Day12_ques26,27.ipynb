```python
from tpot import TPOTClassifier
```

```python
#loading training and testing dataset
train = pd.read_csv("code/data/titanic_train.csv")
```

```python
train.columns
```

```python
train.head()
```

```python
#renaming the traget class name to class
train = train.rename(columns={"survived": "class"})
```

```python
# keeping only relevant columns and dropping high levels of categorical completely
train = train[['class', 'pclass','sex', 'age', 'sibsp', 'parch', 'fare', 'embarked']]
```

```python
#filling the missing values in the dataset
train["age"].fillna(train["age"].median(), inplace=True)
train["embarked"].fillna(train["embarked"].mode()[0], inplace=True)
```

```python
# encoding categorical variables
train['sex'] = train['sex'].map({'male': 0, 'female': 1})
train['embarked'] = train['embarked'].map({'S': 0, 'C': 1, 'Q': 2})
```

```python
#train and target variables
X = train.drop(columns=["class"])
y = train["class"]
```

```python
# splitting the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
# tpot
tpot = TPOTClassifier(generations=5, population_size=20, verbosity=2, random_state=42)
tpot.fit(X_train, y_train)
```

<h2> Keras:
data/pima-indians-diabetes.csv, Binary classification </h2>

```python
# loading the dataset
pima = pd.read_csv('code/data/pima-indians-diabetes.csv', header=None)
```

```python
pima.head()
```

```python
len(pima)
```

```python
X = pima.iloc[:, :-1].values
y = pima.iloc[:, -1].values
```

```python
#feature scaling
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
#splitting the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# defining the model
model = Sequential()
model.add(Dense(12,input_dim=8,activation='relu'))
model.add(Dense(8,activation='relu')) # relu activation for hidden layers
model.add(Dense(1,activation='sigmoid')) # since it is binary classification, we usse sigmoid func
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy']) #compilation of model
```

```python
model.fit(X_train, y_train, epochs=120, batch_size=10, verbose=1)
```

```python
#model evaluation
loss, accuracy = model.evaluate(X_test, y_test)
```
